#Tentativa Correta

OBS: Em todos os resultados é utilizado o mesmo range da máscara de cada cor (Vermelho, Azul, Verde, Amarelo)

Resultado 1


```python
image = cv2.imread( path + 'pare.jfif')
 
## mask of red color
mask1 = cv2.inRange(image, (0, 0, 100), (50, 50,255))

## mask of blue color
mask2 = cv2.inRange(image, (75,0,0), (255,50,50))

## mask of green color
mask3 = cv2.inRange(image, (0,50,0), (75, 255, 25))

## mask of yellow color
mask4 = cv2.inRange(image, (0,100,100), (75, 255, 255))

plt.figure(figsize=[30,22])
plt.subplot(151);plt.imshow(image[:,:,::-1]);plt.axis('off');plt.title("Imagem Original",fontdict={'fontsize': 25});
plt.subplot(152);plt.imshow(mask1, cmap='gray');plt.axis('off');plt.title("Máscara vermelha",fontdict={'fontsize': 25})
plt.subplot(153);plt.imshow(mask2, cmap='gray');plt.axis('off');plt.title("Máscara azul",fontdict={'fontsize': 25})
plt.subplot(154);plt.imshow(mask3, cmap='gray');plt.axis('off');plt.title("Máscara verde",fontdict={'fontsize': 25})
plt.subplot(155);plt.imshow(mask4, cmap='gray');plt.axis('off');plt.title("Máscara amarela",fontdict={'fontsize': 25})
```
